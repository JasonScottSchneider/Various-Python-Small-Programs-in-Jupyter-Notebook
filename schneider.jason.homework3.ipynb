## Program 2

A file loaded into the program by the user should contain names of reviewers and their reviews of different restaurants in
the comma separated format. The first string should be the name of the reviewer followed
by a name of a restaurant and its rating. This Python program computes a similarity score between any two reviewers using Euclidean distances. Example data from a text file is given below:<br>
<br>
<blockquote><u>Pete Wellsworth’s reviews are:</u><br>
'Rooster and Owl': 3.0,<br>
'El Sapo Cuban Social Club': 4.0,<br>
'Punjab Grill': 3.0,<br>
'Shibumi': 5.0,<br>
'Mama Chang': 3.5<br>
<br>
<u>Jay Samuel’s reviews are:</u><br>
'El Sapo Cuban Social Club':4.5,<br>
'Mama Chang':1.0,<br>
'Shibumi':4.0</blockquote><br>
<br>
The Euclidean distance between Jay Samuel and Pete Wellsworth is computed as
follows:<br>
<br>
<blockquote>1) Only consider the restaurants that both have reviewed.<br>
2) Take the difference between corresponding reviews.<br>
3) Sum the square of the differences<br>
4) The square root of the sum of differences is the Euclidean score. The shorter
the distance the closer the two reviewers.</blockquote><br>
<br>
For our example, this would be: (4.0 – 4.5)2 + (3.5 – 1.0)2 + (5.0 – 4.0)2 = 0.25 +
6.25 + 1 = 7.5<br>
Euclidean distance is Square Root of 7.5 = 2.7386<br>
<br>
This program should provide the following:<br>
<blockquote>1) Ability to read in a user provided filename which contains restaurant reviews
in the comma separated format similar to the given file restaurant_reviews.txt<br>
2) Functionality to compute the similarity between two reviewers provided by the
user<br>
3) Functionality to compute the similarity between one user provided reviewer and
all other reviewers in the database</blockquote><br>
<br>
A sample run for the two required functionalities is given below (user inputs are in
bold):<br>

What do you want to do? Input 1 for similarity between two reviewers, or
Input 2 for similarity between one reviewer and all others in the database or 3
to quit: <b>3</b><br>
Goodbye!</blockquote><br>

```python
import math
import re

x=input('Give the name of the restaurant reviews file: ')
try:   
    #reads all the lines in the file into a list called lines
    #with-statements already close files unlike using open and readLines-statements
    with open(x) as filehandle:
        lines=filehandle.readlines()

    #Removes empty lines
    ndata = []
    for line in lines:
        line = line.strip()
        if line:
            ndata.append(line)        

    #Gets the name of the reviewers in a list
    listedLines=[]
    for eachLine in ndata:
        y=eachLine.split(",")
        listedLines.append(y[0])

    #Gets the values of the reviewers in a list and removes \n
    data=[]
    for everyLine in lines:
        everyLine=everyLine.rstrip("\n")
        #if len(everyLine.split()) == 0:
            #continue
        everyLine=everyLine.split(",",maxsplit=1)[1:]
        data.append(everyLine)
    
    #Makes sublists from the list of the reviewers' values;divides each sublist into listed strings as prep to make a nested dictionary later
    allData=[]
    for everyLine in data:
        for everyword in everyLine:
            everyword=everyword.split(",")
            allData.append(everyword)

    #Convert sublists of strings into a dictionary of restuarants as keys and scores as values
    bigData=[]
    for everylist in allData:
        everylist={everylist[i]: everylist[i + 1] for i in range(0, len(everylist), 2)}
        bigData.append(everylist)         

    def get_dic_from_two_lists(keys, values):
        return { keys[i] : values[i] for i in range(len(keys)) }        
    #Makes a dictionary from two lists by nesting dictionaries as values connected to our reviewers who are the upper level keys
    dataDict=get_dic_from_two_lists(listedLines, bigData)

    #Sets up dictionaries to be compared by only the similar keys and values between the two compared
    def compare_dicts(vec1,vec2):
        updates = {}
        for k2, v2 in vec1.items():
            if k2 in vec2:    
                updates.update({k2 : v2})
        return updates

    #Puts the keys and values between the two dictionaries compared in the same order to each other
    def sortThedicts(x):
        dictList=[]
        for key, value in x.items():
            dictList.append([key, value])
        dictList.sort()
        dictList=dict(dictList)
        return dictList

    #Calculates the Euclidean distance
    def euclidean_distance(vec1,vec2):
            distance = 0
            for i in range(len(vec1)):
                x = float(vec1[i]); y = float(vec2[i])
                distance += (x-y)**2
            return math.sqrt(distance)
    
    #Main Program
    while True:
        try:
            choice=int(input('What do you want to do? Input 1 for similarity between two reviewers, or input 2 for similarity between one reviewer and all others in the database, or input 3 to quit: '))
        except ValueError:
            print('You entered something else other than 1, 2, or 3. Try again.')
            continue
        #Compares one reviewer with another single reviewer
        if choice is 1:
            reviewer1=str(input('Provide Reviewer1 name: '))
            try:    
                dataDict1 = dataDict[reviewer1]
            except KeyError:
                print('You entered something else other than a name of a reviewer. Start over.')
                continue                          
            reviewer2=str(input('Provide Reviewer2 name: '))
            try:               
                dataDict2 = dataDict[reviewer2]
            except KeyError:
                print('You entered something else other than a name of a reviewer. Start over.')
                continue 
            
            #Filters compared dictionaries for only same keys between the compared dictionaries
            ndataDict1 = compare_dicts(dataDict1,dataDict2)
            ndataDict2 = compare_dicts(dataDict2,dataDict1)
            
            #Sorts the compared dictionaries
            ndataDict1=sortThedicts(ndataDict1)
            ndataDict2=sortThedicts(ndataDict2)

            #Extracts just the values from the nested dictionaries into lists
            list1 = list(ndataDict1.values())
            list2 = list(ndataDict2.values())

            #Calculates distance
            sim = euclidean_distance(list1,list2)

            print('The similarity score between ',reviewer1, 'and ',reviewer2,'is: ',sim)  

        #Compares one reviewer with all other reviewers
        elif choice is 2:
            reviewer1=str(input('Provide Reviewer1 name: '))
            try:                
                dataDict1 = dataDict[reviewer1]
            except KeyError:
                print('You entered something else other than a name of a reviewer. Start over.')
                continue 
                
            print('The Similarity Scores are: ')

            #Excludes the selected reviewer to be compared with all other reviewers so that the selected reviewer is not compared with him or herself
            try:
                del dataDict[reviewer1]
            except KeyError:
                pass

            stringKey=()
            for key, value in dataDict.items():
                dataDict2=dataDict[key]#All other reviewers to be compared to the selected reviewer
                stringKey=str(key)

                ndataDict1 = compare_dicts(dataDict1,dataDict2)
                ndataDict2 = compare_dicts(dataDict2,dataDict1)

                ndataDict1=sortThedicts(ndataDict1)
                ndataDict2=sortThedicts(ndataDict2)

                list1 = list(ndataDict1.values())
                list2 = list(ndataDict2.values())

                sim = euclidean_distance(list1,list2)

                print(reviewer1, stringKey, sim)
        
        #Stops and quits the program
        elif choice is 3:
            print('Goodbye!')
            break     
#Print an error in no document found at the input
except IOError as error:
    print('The file location entered cannot be opened: ',x)
```
