# Program 2
## Please read the following for instructions and examples.
&emsp;This is a program that reads in three or four integers and displays the following:<br>
a. the average of the numbers (two decimal places using a float)<br>
b. the maximum of the numbers<br>
c. the minimum of the numbers<br>
d. the median of the numbers (If there are two numbers that qualify for the median, both will be calculated and displayed from the mean of the two numbers.)<br>
<br>
The statistical functions for mean, median, max, and min found in Python are not used and, instead, are replaced with manually written definitions in the cell below.


```python
#2
def ssum(lst):
    accumulator = 0
    for element in lst:
        accumulator += element
    return accumulator

def calculate_mean(numbers):
    s = ssum(numbers)#total of numbers in a tuple
    N = len(numbers)#total number of inputs, not the value of the inputs, in a tuple
    #Calculate the mean
    mean = s/N#total of the value entered for inputs are divided by the total number of inputs entered, not the value of inputs
    return mean

def calculate_median(numbers):
    N = len(numbers)#total number of inputs in a list
    numbers.sort()
    #we cannot use tuples with the sort function because it must alter the list; 
    #therefore, we must use a list, not a tuple.
    
    #Find the median
    if N % 2 == 0:#the remainder is given from the total number of inputs squared
        #if N is even
        m1 = N/2
        m2 = (N/2) + 1
        #Convert to integer, match position
        m1 = int(m1) - 1
        m2 = int(m2) - 1
        median = (numbers[m1] + numbers[m2])/2
    else:
        m = (N+1)/2
        #Convert to integer, match position
        m = int(m) - 1
        median = numbers[m]       
    return median

def my_min_function(somelist):
    min_value = None#create an empty bin
    for value in somelist:#create a conditional for-statement using Boolean operators to determine the minimum value.
        if not min_value:
            min_value = value
        elif value < min_value:
            min_value = value
    return min_value #fill that bin labeled min_value

def my_max_function(somelist):
    max_value = None#create an empty bin
    for value in somelist:#create a conditional for-statement using Boolean operators to determine the maximum value.
        if not max_value:
            max_value = value
        elif value > max_value:
            max_value = value
    return max_value #fill that bin labeled max_value
```

# Program 3
## Instructions and Examples
&emsp;Program three will request the user to enter a positive integer from the user. Program three will determine whether the integer entered by the user is a prime number with the message, "The number you picked is a prime number" or composite number with the message, "The number you picked is a composite number." If the user enters neither a prime nor composite number, the program will print the message, "The number you picked is not a prime or composite number." The cell below contains the manually written definitions for the functions to detect prime or composite numbers. The user must load the functions before executing the program. 

```python
#3
def isPrime(n) : 
  
    # Corner cases
    if (n <= 1) : 
        return False
    if (n <= 3) : 
        return True
  
    # This is checked so that we can skip  
    # middle five numbers in below loop. 
    if (n % 2 == 0 or n % 3 == 0) : 
        return False
  
    i = 5
    while(i * i <= n) : #when 5 times 5 is less than or equal to a number
        if (n % i == 0 or n % (i + 2) == 0) :#if equal to zero, then return false 
            return False
        i = i + 6
    return True

def isComposite(n): 
  
    # Corner cases 
    if (n <= 1): #if number is less than or equal to 1, return false
        return False
    if (n <= 3): #if number is less than or equal to 3, return false
        return False
  
    # This is checked so that we can skip  
    # middle five numbers in below loop. 
    if (n % 2 == 0 or n % 3 == 0): 
        return True
    i = 5
    while(i * i <= n): #when 5 times 5 is less than or equal to a number
          
        if (n % i == 0 or n % (i + 2) == 0): #if equal to zero, then return false
            return True
        i = i + 6       
    return False    
```

At the end of each run, the user should be prompted to ask whether the user wants to continue with lower case yes or no as much as the user desires. If the user enters anything else, the user gets one more chance to enter yes or no before the program stops and bids the user goodbye.<br>

```python
#5
while True:
    try:
        x1 = float(input("Enter an integer or float for x1: ")) #use float to take both integers and floats
        y1 = float(input("Enter an integer or float for y1: "))
        x2 = float(input("Enter an integer or float for x2: "))
        y2 = float(input("Enter an integer or float for y2: "))
        x = float(input("Enter an integer or float for x: "))
        y = float(input("Enter an integer or float for y: "))
    except ValueError: #catch entered character strings
        print("You entered a character string where there should not be one. Goodbye.")
        break

    if(x >= x1 and x <= x2) and (y >= y1 and y <= y2): #check whether x and y fit within the parameters of a rectangle defined
        print(True) #the Boolean turns out true
    else:
        print(False) #the Boolean turns out false
    a = input("Enter yes/no in lower case to continue: ") #prompt the user to play again     
    if a=="yes":
        continue
    elif a=="no":
        print('Goodbye.')
        break
    elif a!='yes' or a!='no': #catch strings entered other than yes or no from the user
        print("Enter either yes/no in lower case. If you do not enter yes/no, the program will terminate.")
        a = input("Last chance, enter yes/no to continue: ") #the user has been warned with a last shot
        if a=="yes":
            continue
        elif a=="no":
            print('Goodbye.')
            break
        elif a!='yes' or a!='no':
            print('I told you so if you entered anything other than yes/no. Goodbye.')
            break
```
